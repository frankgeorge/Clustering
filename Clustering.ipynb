```
import numpy as np
import pandas as pd
```

```
from kmodes.kprototypes import KPrototypes
```

```
df = pd.read_csv('/content/drive/MyDrive/marketing_cva_f.csv')
```

```
df
```

```
df=df.drop(['Customer', 'Vehicle_Class','avg_vehicle_age', 'months_last_claim', 'Total_Claim_Amount'],axis=1)
```

```
df
```

```
array=df.values
```

```
array[:,1] = array[:, 1].astype(float)
array[:,3] = array[:, 1].astype(float)
array[:,5] = array[:, 1].astype(float)
array[:,6] = array[:, 1].astype(float)
```

```
array
```

```
kproto = KPrototypes(n_clusters=3, verbose=2, max_iter=20)
clusters = kproto.fit_predict(array, categorical=[0,2,4])
```

```
print(kproto.cluster_centroids_)
```

```
cluster_dict=[]
for c in clusters:
  cluster_dict.append(c)
```

```
cluster_dict
```

```
df['cluster']=cluster_dict
```
